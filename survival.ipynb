```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset from the provided URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Basic EDA: Data Overview
df_info = df.info()
df_description = df.describe()

# Plot 1: Survival rate of passengers
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Survived', palette="Set2")
plt.title("Survival Rate of Passengers")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# Plot 2: Average age of passengers
plt.figure(figsize=(6, 4))
sns.histplot(df['Age'].dropna(), kde=True, color='blue')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Plot 3: Proportion of male and female passengers
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sex', palette="Set2")
plt.title("Proportion of Male and Female Passengers")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()

# Plot 4: Survival rate by class
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Pclass', y='Survived', palette="Set2")
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()

# Plot 5: Distribution of the fare paid by the passengers
plt.figure(figsize=(6, 4))
sns.histplot(df['Fare'], kde=True, color='green')
plt.title("Distribution of Fare Paid by Passengers")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```
